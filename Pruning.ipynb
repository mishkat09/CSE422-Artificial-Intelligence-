```python
import math as mat
import random as rand

new_id = input("Enter your student ID: ")
new_list = list(new_id)

for i in range(len(new_list)):
    if new_list[i] == "0":
        new_list[i] = 8
list_of_id = []
for i in range(len(new_list)):
    list_of_id.append(int(new_list[i]))

minPoints = list_of_id[4] # for minimum limit points

# for goal and maximum points

integer = list_of_id[len(list_of_id) - 2: len(list_of_id)]
tmp= integer[0]
integer[0] = integer[1]
integer[1] = tmp
maxP = []
for i in range(len(integer)):
    maxP.append(integer[i])
maximum = list(map(str, maxP))
string = ""
for i in range(len(maxP)):
    string += maximum[i]
goal = int(string)   # goal point

if goal%2 == 0:  # for maximum limit point
    maxPoints = int(goal * 1.5)
else: maxPoints = int((goal * 1.5) + 0.5)

depth = 3
alpha = -mat.inf
beta = mat.inf

winCount = 0

randomList = []
for i in range(len(new_id)):
    randomList.append(rand.randint(minPoints, maxPoints))

def alphapruning(depth, node, flag, mainList, alpha, beta, branch): # alphaBetaPruning
    if depth == 0:
        return mainList[node]
    if flag == True:
        attacker = alpha
        for index in range(0, branch):
            points = alphapruning(depth - 1, node * branch + index, False, mainList, alpha, beta, 2)
            attacker = max(alpha, points)
            alpha = max(alpha, attacker)
            if beta <= alpha:
                break
        return attacker
    else:
        defender = beta
        for index in range(0, branch):
            points = alphapruning(depth - 1, node * branch + index, True, mainList, alpha, beta, 2)
            defender = min(beta, points)
            beta = min(beta, defender)
            if beta <= alpha:
                break
        return defender


points = alphapruning(depth, 0, True, randomList, alpha, beta, 2)

if points >= goal:
    winString = "The Winner is Optimus Prime"
else:
    winString = "The winner is Megatron"
print("Generated 8 random points between the minimum and maximum point \nlimits: ", randomList)
print("Total points to win:", goal)
print("Achieved point by applying alpha-beta pruning =", points)
print(winString)

# For task 2

pointList = []
shuffle = list_of_id[3]
for i in range(shuffle):
    rand.shuffle(randomList)
    points = alphapruning(depth, 0, True, randomList, alpha, beta, 2)
    pointList.append(points)
    if points >= goal:
        winCount = winCount + 1

maximumPointinList = max(pointList)

print("")
print("After the shuffle: ")
print("List of all points values from each shuffle:", pointList)
print("The maximum value of all shuffles:",maximumPointinList)
print(f"Won {winCount} times out of {shuffle} number of shuffles")
```
