```python
import random


class Transaction:
    def __init__(self, amount, is_lend):
        self.amount = amount
        self.is_lend = is_lend

    def __repr__(self):
        return f"{'l' if self.is_lend else 'd'} {self.amount}"


def fitness_function(register):

    return sum(t.amount for t in register if t.amount != 0)


def crossover(parent1, parent2):

    crossover_point = random.randint(0, len(parent1)-1)
    child = parent1[:crossover_point] + parent2[crossover_point:]

    return child


def mutation(register, probability):

    for i in range(len(register)):
        if random.random() < probability:
            register[i].amount *= -1

    return register

# def genetic_algorithm(transactions, population_size, crossover_rate, mutation_rate, max_generations):

#     register = [Transaction(int(amount), is_lend=='l') for is_lend, amount in transactions]


#     population = [mutation(register[:], mutation_rate) for i in range(population_size)]

#     fitness_scores = [fitness_function(p) for p in population]

def genetic_algorithm(transactions, population_size, crossover_rate, mutation_rate, max_generations):

    register = [Transaction(int(amount), is_lend=='l') for is_lend, amount in transactions]


    population = [mutation(register[:], mutation_rate) for i in range(population_size)]

    fitness_scores = [fitness_function(p) for p in population]


    for generation in range(max_generations):

        for i, p in enumerate(population):
            if fitness_scores[i] == 0:
 
                binary_string = ''.join(['1' if t.amount < 0 else '0' for t in p])
                return binary_string

        if sum(fitness_scores) <= 0:
            continue

        parents = random.choices(population, weights=fitness_scores, k=population_size)

        new_population = []
        for i in range(0, population_size, 2):
            parent1, parent2 = parents[i], parents[i+1]
            if random.random() < crossover_rate:
                child1 = crossover(parent1, parent2)
                child2 = crossover(parent2, parent1)
            else:
                child1, child2 = parent1[:], parent2[:]
            child1 = mutation(child1, mutation_rate)
            child2 = mutation(child2, mutation_rate)
            new_population.extend([child1, child2])


        new_fitness_scores = [fitness_function(p) for p in new_population]

        population = new_population
        fitness_scores = new_fitness_scores


    return -1

n = int(input())
transactions = [input().split() for i in range(n)]


result = genetic_algorithm(transactions, population_size=100, crossover_rate=0.8, mutation_rate=0.1, max_generations=100)


print(result)
```
